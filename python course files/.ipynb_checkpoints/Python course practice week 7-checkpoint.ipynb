### 3.直方图
- 表示数据离散统计分布
- 参数bins=n，就是把数据范围（最小值到最大值）等分为n个区间，区间高度就是数据量
- 参数alpha控制透明度

```python
df = pd.DataFrame({"a": np.random.randn(1000)+1,
                  "b": np.random.randn(1000),
                  "c": np.random.randn(1000)-1},
                 columns=list("abc"))
# a为期望为1方差为1的标准正态分布，b期望为0，c期望为-1，其他相同
df.plot.hist(bins=20, alpha=0.5)
```

```python
df.plot.hist(bins=100, alpha=0.5)
```

```python
df.plot.hist(bins=20, alpha=0.9)
```

### 4.箱型图/盒型图/箱线图
- Series.plot.box()或DataFrame.plot.box()或DataFrame.boxplot()
- 用五个值体现数据的分布（max，75%，50%，25%，min）
- 绿线：中位数
- 上下蓝线：上四分位数（Q3），下四分位数（Q1）
- IQR = Q3 - Q1（interquartile range）
- 上下黑线:符合合理范围内的最大最小值
- 合理范围：最大值范围：Q3+1.5*IQR，最小值范围：Q1-1.5*IQR
- 异常值：超出最大最小值范围的数据标注异常值

```python
df = pd.DataFrame(np.random.randn(100, 5), columns=list("abcde"))
df.plot.box()
```

```python
df.boxplot()
```

### 3. 完整的绘制程序

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *

# 定义数据
x = np.arange(0., 10, 0.2)
y1 = np.cos(x)
y2 = np.sin(x)
y3 = np.sqrt(x)

# 绘制3条函数曲线
plt.rcParams["figure.figsize"] = (12,8)
plt.plot(x,y1,color="blue",linewidth=4,linestyle="-.",marker=".",label=r"$y = cos{x}$")
plt.plot(x,y2,color="green",linewidth=1.5,linestyle="--",marker="*",label=r"$y = sin{x}$")
plt.plot(x,y3,color="m",linewidth=1.5,linestyle="-",marker="x",label=r"$y = \sqrt{x}$")

# 去框线
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")

# 移动x，y轴
ax.xaxis.set_ticks_position('bottom')
ax.spines["bottom"].set_position(("data",0))
ax.yaxis.set_ticks_position('left')
ax.spines["left"].set_position(('data',0))

# 设置x，y轴的取值范围
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.ylim(-1.5, 4.0)

# 设置x，y轴的刻度值
plt.xticks([2,4,6,8,10],[r'two', r'four', r'six', r'8', r'10', ])
plt.yticks([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0],
          [r'-1.0', r'0.0', r'1.0', r'2.0', r'3.0', r'4.0', ])

# 添加文字
# plt.text(0.8, 0.8, r'$x \in[0.0, \ 10.0]$', color='k', fontsize=15)
# plt.text(0.8, 0.89 r'$x \in[-1.0, \ 4.0]$', color='k', fontsize=15)

# 特殊点添加注解
plt.scatter([8,], [np.sqrt(8),], 50, color='m') # 使用散点图放大当前点
plt.annotate(r'$2\sqrt{2}$', xy=(8, np.sqrt(8)), xytext=(8.5,2.2), fontsize=16,color='#090909')

# 设置标题，x轴，y轴
plt.title(r'$the\ function\ figure\ of\ cos(),\ sin()\ and\ sqrt()$', fontsize=20)
plt.xlabel(r'$the\ input\ value\ of\ x$', fontsize=15, labelpad=20)
plt.ylabel(r'$y = f(x)$', fontsize=15, labelpad=20)

# 设置图例及位置
plt.legend(loc='upper right')

# 显示网格线
plt.grid(True)

# 显示绘图
plt.show()
```

### 拟合不同模型
- 有时候线性拟合效果并不好，我们就需要拟合其他模型

```python
anscombe = sns.load_dataset("anscombe")
anscombe.head()
```

```python
anscombe.query("dataset == 'I'")
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"), ci=None,
           scatter_kws={"s":50})
# 数据点size取50
```

```python
anscombe.query("dataset == 'II'")
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"), ci=None,
           scatter_kws={"s":50})
# 会发现线性拟合明显效果不好
```

#### 尝试高阶拟合
- order = n  n阶多项式拟合

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),order=2 ,ci=None,
           scatter_kws={"s":50})
```

```python
anscombe.query("dataset == 'III'")
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),order=1 ,ci=None,
           scatter_kws={"s":50})
# 发现异常值对拟合影响很大
```

#### 处理异常值
- robust=True

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),order=1 ,ci=None,
           robust=True,
          scatter_kws={"s":50})
```

#### 如何评价拟合效果：残差曲线
- residual plot
- residplot()
- 拟合的比较好就会拿到白噪声的分布 $N(0,\delta^2)$
- 白噪声检测在时间序列分析课程中，不赘述，详情自行学习，检测p值越大，表明越有可能剩下的序列是一个纯随机序列

```python
sns.residplot(x="x",y="y",data=anscombe.query("dataset == 'I'"), scatter_kws={'s':50})
```

像下面这个残差序列明显还有一个二次关系，说明原序列没有拟合好

```python
sns.residplot(x="x",y="y",data=anscombe.query("dataset == 'II'"), scatter_kws={'s':50})
```

### 5. 同时绘制多张图

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks")
%matplotlib inline
```

#### 用column分组，并用map函数绘制想要的图

```python
tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="time")
```

```python
tips.head()
```

```python
g = sns.FacetGrid(tips, col="time")
g.map(plt.hist, "tip")
```

```python
g = sns.FacetGrid(tips, col="sex", hue="smoker")
g.map(plt.scatter, "total_bill", "tip", alpha=.7)
g.add_legend()
```

```python
g = sns.FacetGrid(tips, col="day", height=4, aspect=.5)
g.map(sns.barplot, "sex", "total_bill")
# height纵轴高度，aspect横纵轴长度比例
```

```python
ordered_days = tips.day.value_counts().index
g = sns.FacetGrid(tips, row="day", height=1.7, row_order=ordered_days, aspect=4)
g.map(sns.distplot, "total_bill", hist=False ,rug=True)
# row_order控制多行图的顺序，这里是按星期几出现次数排序
```

```python
attend = sns.load_dataset("attention").query("subject<=12")
attend.head()
```

```python
g = sns.FacetGrid(attend, col="subject", col_wrap=4, height=2, ylim=(0,10)) 
g.map(sns.pointplot, "solutions","score", color=".3", ci=None)
# col_wrap控制几个图换行显示 ylim控制纵轴范围
```

```python
g = sns.FacetGrid(tips, col="smoker", margin_titles=True, height=4)
g.map(plt.scatter, "total_bill", "tip", color="#338844", edgecolor="white", s=50, lw=1)
for ax in g.axes.flat:
    ax.plot((0,50),(0,.2*50), c=".2", ls="--")
g.set(xlim=(0,60), ylim=(0,14))
# margin_titles页边标题
```
